```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Створення часової осі
t_start = 0
t_end = 1
dt = 0.001
t = np.arange(t_start, t_end, dt)

# 2. Імпульсний сигнал (прямокутний імпульс)
impulse = np.where((t >= 0.2) & (t <= 0.4), 1.0, 0.0)

# 3. Періодичний сигнал (синусоїда)
A = 1.0
f = 5.0
phi = 0.0
sine_signal = A * np.sin(2 * np.pi * f * t + phi)

# 4. Випадковий сигнал
random_signal = np.random.uniform(-1, 1, len(t))

# 5. Побудова графіків
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, impulse)
plt.title("Імпульсний сигнал")
plt.xlabel("Час, с")
plt.ylabel("Амплітуда")
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, sine_signal)
plt.title("Періодичний сигнал (синусоїда)")
plt.xlabel("Час, с")
plt.ylabel("Амплітуда")
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, random_signal)
plt.title("Випадковий сигнал")
plt.xlabel("Час, с")
plt.ylabel("Амплітуда")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Характеристики змодельованих сигналів:

- прямокутний імпульс: на графіку видно чітке «вікно» від 0.2 до 0.4 секунди. Сигнал має лише два стани: нуль (відсутність сигналу) та одиницю (наявність). Це детермінований сигнал, обмежений у часі.
- періодичний сигнал (синусоїда) – плавна лінія, що повторюється. При частоті 5 Гц – 5 повних циклів за 1 секунду. Приклад неперервного за своєю природою сигналу.
- випадковий сигнал (шум): графік виглядає як хаотичні стрибки амплітуди в межах від -1 до 1. Не має закономірності чи періодичності.

```python
# 1. Імпульсний сигнал (прямокутний імпульс)
impulse = np.where((t >= 0.2) & (t <= 0.4), 2.0, 0.0)

# 2. Періодичний сигнал (синусоїда)
A = 1.0
f1, f2 = 10.0, 20.0
phi = 0.0
sine_signal_10Hz = A * np.sin(2 * np.pi * f1 * t + phi)
sine_signal_20Hz = A * np.sin(2 * np.pi * f2 * t + phi)

# 3. Випадковий сигнал
random_signal1 = np.random.uniform(-1, 1, len(t))
random_signal2 = np.random.uniform(-1, 1, len(t))

signals = {'Імпульсний сигнал': impulse,
           "Періодичний сигнал (синусоїда, частота 10 Гц)": sine_signal_10Hz,
           "Періодичний сигнал (синусоїда, частота 20 Гц)": sine_signal_20Hz,
           "Випадковий сигнал 1": random_signal1,
           "Випадковий сигнал 2": random_signal2}

# Побудова графіків
plt.figure(figsize=(10, 12))

for i, (name, sig) in enumerate(signals.items(), start=1):
    plt.subplot(len(signals), 1, i)
    plt.plot(t, sig)
    plt.title(name)
    plt.xlabel("Час, с")
    plt.ylabel("Амплітуда")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

Характеристики змодельованих сигналів:
- прямокутний імпульс: при зміні амплітуди візуально змінюється лише висота «сходинки» по осі У, при зміні тривалості – змінюється ширина «сходинки».
- періодичний сигнал (синусоїда): при збільшенні частоти до 10 Гц та 20 Гц період коливань скорочується, цикли стають щільнішими. При збільшенні амплітуди А – змінюється висота графіка по осі У; зі зміною значення phi відбувається зміщення праворуч при phi < 0 або ліворуч для phi > 0.
- випадковий сигнал (шум): порівняння різних випадкових сигналів показує, що хоча вони мають однакові межі амплітуди, їхня форма ніколи не зберігається.


```python
clean_signal = sine_signal_10Hz
noise = np.random.uniform(-0.3, 0.3, len(t))
noisy_signal = clean_signal + noise

plt.figure(figsize=(10, 5))
plt.plot(t, clean_signal, label="Чистий сигнал")
plt.plot(t, noisy_signal, label="Сигнал із шумом", alpha=0.5)
plt.xlabel("Час, с")
plt.ylabel("Амплітуда")
plt.title("Періодичний сигнал (синусоїда, частота 10 Гц) із шумом")
plt.legend()
plt.grid(True)
plt.show()
```

При додаванні випадкового шуму до чистої синусоїди отримуємо модель реального сигналу. Чиста лінія стає «зубчатою». Це демонструє, як зовнішні завади можуть спотворювати корисну інформацію, хоча загальна форма синусоїди (її тренд) залишається впізнаваною.

```python
# Імпульсний сигнал (прямокутний імпульс)
T = 1.0   # тривалість
steps = [20, 50, 200]  # кількість точок дискретизації

plt.figure(figsize=(12, 6))

for i, N in enumerate(steps, start=1):
    t = np.linspace(0, T, N)
    impulse = np.where((t >= 0.2) & (t <= 0.4), 2.0, 0.0)  # прямокутний імпульс

    plt.subplot(len(steps), 1, i)
    plt.plot(t, impulse, marker='o')
    plt.title(f"Прямокутний імпульс при {N} точках дискретизації")
    plt.xlabel("Час, с")
    plt.ylabel("Амплітуда")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Оригінальний безперервний сигнал (синусоїда)
f = 5.0  # частота
T = 1.0  # тривалість
phi = 0.0

# Варіанти дискретизації
steps = [20, 50, 200]  # кількість точок

plt.figure(figsize=(12, 6))

for i, N in enumerate(steps, start=1):
    t = np.linspace(0, T, N)  # різна кількість точок
    signal = np.sin(2 * np.pi * f * t + phi)

    plt.subplot(len(steps), 1, i)
    plt.plot(t, signal, marker='o')
    plt.title(f"Синусоїда при {N} точках дискретизації")
    plt.xlabel("Час, с")
    plt.ylabel("Амплітуда")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

Експерименти зі різною кількістю точок (N = 20, 50, 200) наочно показують суть дискретизації:

-	при N = 20: сигнал виглядає ламаним. Прямокутний імпульс може змінити свої краї, якщо точки дискретизації не потраплять точно на моменти 0.2 с або 0.4 с.
-	при N = 200: візуально сигнал стає плавним і майже не відрізняється від аналогового.

Висновок: чим більше точок дискретизації (менше крок), тим точніша цифрова копія відповідає оригіналу, але тим більше пам’яті потрібно для зберігання таких даних.


**Висновки**

Було реалізовано три основні класи сигналів: імпульсні (локалізовані в часі), періодичні (гармонічні коливання) та випадкові (стохастичні процеси).

Основні властивості:

-	імпульсні сигнали характеризуються тривалістю та амплітудою;
-	періодичні – частотою, періодом та фазою;
-	випадкові – діапазоном значень.

Відмінності між сигналами:

-	імпульсний має чіткій початок та кінець;
-	періодичний сигнал передбачуваний і повторюється нескінченно;
-	випадковий не має функціональної залежності від часу, його значення неможливо передбачити.

Аналіз отриманих результатів:

-	зміна амплітуди впливає на потужність сигналу, а зміна частоти на швидкість передачі змін у часі;
-	на графіку синусоїди з шумом видно, що чим більша амплітуда корисного сигналу відносно шуму, тим легше його розпізнати (співвідношення сигнал/шум);
-	дослідження кроку дискретизації підтверджує, що для якісного відтворення сигналу частота дискретизації повинна бути значно вищою за частоту самого сигналу (відповідно до теореми Котельникова).

Можливості Python: для моделювання було використано бібліотеку NumPy (генерування масивів точок, обчислення тригонометричних функцій, створення випадкових чисел через random)  та бібліотеку Matplоtlib (візуалізація даних, створення графіків, налаштування сітки та легенди).

Здобуті практичні навички:

-	програмного генерування часових рядів;
-	розуміння впливу параметрів дискретизації на візуальну та математичну точність сигналу;
-	вміння комбінувати сигнали (накладання шуму на корисний сигнал);
-	досвід аналізу графічної інформації та порівняння теоретичних моделей з їх цифровими копіями.
